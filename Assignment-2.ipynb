```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
var=urllib.request.urlopen(url).read()
soup=BeautifulSoup(var,'xml')
print(soup.prettify())
```

```python
table=soup.find('table')
column_names = ['Postalcode','Borough','Neighborhood']
toronto = pd.DataFrame(columns = column_names)
for tr_val in table.find_all('tr'):
    row=[]
    for td_val in tr_val.find_all('td'):
        row.append(td_val.text.strip())
    if len(row)==3:
        toronto.loc[len(toronto)]=row
toronto.head()
        

```

Using only with Borough != Not Assigned

```python
toronto=toronto[toronto['Borough']!='Not assigned']
toronto
```

```python
toronto.loc[toronto['Neighborhood']=='Not Assigned','Neighborhood']=toronto['Borough']
toronto
```
